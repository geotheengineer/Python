# 3D vector ploting

```python
from numpy import *
import pylab as p
import mpl_toolkits.mplot3d.axes3d as p3
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Arc
from matplotlib.patches import FancyArrowPatch
from mpl_toolkits.mplot3d import proj3d
#from mpl_toolkits.mplot3d import Axes3D
import mpl_toolkits.mplot3d.art3d as art3d
#-------------------------------------------------------------------
class force:
    def __init__self(self, mag, dr, dz, dh, fx, fy,fz):
        self.mag = mag 
        self.dr = dr  # dr = theta   NOTE that the y-axis is ZERO (NOT x-axis!)
        self.dz = dz  # dz = phi
        self.dh = dh  # dh = the projection onto the x-y plane, to calc arctan(phi)
        self.fx = fx 
        self.fx = fy 
        self.fz = fz
    description = "a 3D force, given by its magnitude and angle in degrees"
    def resolve(self):
        self.dr = (self.dr*math.pi)/180 #convert degrees to radians for trig functs
        self.dz = (self.dz*math.pi)/180
        self.dh = self.mag*cos(self.dz)
        self.fx = self.mag*cos(self.dz)*sin(self.dr)
        self.fy = self.mag*cos(self.dz)*cos(self.dr)
        self.fz = self.mag*sin(self.dz)
        
#--------------------Vectors to be summed:---------------------
        """
        NEED to setup c++-style vector for forces, & iterate through, 
        summing RES with a FOR-loop.
        AND also enter the info into the Arrow command, so all will be drawn
        --Do BOTH in one loop?  Simplifying naming  . .
        
        Vector of only the Mag & Dir info, then for-loop to populate, resolve(),
        declare arrow for each, and   res.fx += [i].fx . . & so forth . .         
        
        --ALSO:  a more user-friendly way to enter the Force info ?
        SEPARATE & cout to command line to enter vectors?  
          (for testing, just auto-fill) - run with 'interactive' option off/on
         Read from input, and read fron FILE !!
        """
        """  Animate in FOR-loop as # of vectors increases & are summed ? """
f = force()
f.mag = 100
f.dr = 0
f.dz = 25
f.resolve()

g = force()
g.mag = 100
g.dr = 120
g.dz = 15
g.resolve()

h = force()
h.mag = 100
h.dr = 240
h.dz = 35
h.resolve()
#--------------------Resultant vector calculations:--------------
res = force()
res.fx = (f.fx + g.fx + h.fx)
res.fy = (f.fy + g.fy + h.fy)
res.fz = (f.fz + g.fz + h.fz)
res.mag = sqrt(res.fx**2 + res.fy**2 + res.fz**2)
res.dr = math.atan(res.fy/res.fx)
res.dh = sqrt(res.fx**2 + res.fy**2)
res.dz = math.atan(res.fz/res.dh)
#------------------------------------------------------------
fig = p.figure() 
ax = p3.Axes3D(fig)
ax.set_xlim3d(-(g.mag), (g.mag))  # Axis - control:  Now set for vector g
ax.set_ylim3d(-(g.mag), (g.mag))
ax.set_zlim3d(0, (g.mag))
#--------------------------------------------------------------

class Arrow3D(FancyArrowPatch):
    def __init__(self, xs, ys, zs, *args, **kwargs):
        FancyArrowPatch.__init__(self, (0,0), (0,0), *args, **kwargs)
        self._verts3d = xs, ys, zs

    def draw(self, renderer):
        xs3d, ys3d, zs3d = self._verts3d
        xs, ys, zs = proj3d.proj_transform(xs3d, ys3d, zs3d, renderer.M)
        self.set_positions((xs[0],ys[0]),(xs[1],ys[1]))
        FancyArrowPatch.draw(self, renderer)

a = Arrow3D([0,res.fx],[0,res.fy],[0,res.fz], mutation_scale=20, \
    lw=1, arrowstyle="-|>", color="r", label='R')
b = Arrow3D([0,f.fx],[0,f.fy],[0,f.fz], mutation_scale=20, lw=1, \
     arrowstyle="-|>", color="k", label='f')
c = Arrow3D([0,g.fx],[0,g.fy],[0,g.fz], mutation_scale=20, lw=1, \
    arrowstyle="-|>", color="k", label='g')
d = Arrow3D([0,h.fx],[0,h.fy],[0,h.fz], mutation_scale=20, lw=1, \
    arrowstyle="-|>", color="k", label='h')

ax.add_artist(a)
ax.add_artist(b)
ax.add_artist(c)
ax.add_artist(d)
#==========================================
p = Circle((0, 0), g.mag*.4, edgecolor='0.6', facecolor='none', linewidth=0.5)
ax.add_patch(p)
art3d.pathpatch_2d_to_3d(p, z=0, zdir="z")

ax.text(-g.mag*.4, 0, 0, "90", color='0.6')
ax.text(0, -g.mag*.4, 0, "180", color='0.6')
ax.text(0, g.mag*.4, 0, "0", color='0.6')
ax.text(0, 0, g.mag*.4, "90", color='0.6')
ax.text(0, 0, -g.mag*.4, "-90", color='0.6')

ax.text2D(0.01, 0.96, 'Resultant:   %.2f N,    %.2f horiz,    %.2f vert ' 
    %( res.mag , (res.dr*180/math.pi), (res.dz*180/math.pi) ), 
    color='r', transform=ax.transAxes)
k = Circle((0, 0), g.mag*.4, edgecolor='0.6', facecolor='none', linewidth=0.5)
ax.add_patch(k)
art3d.pathpatch_2d_to_3d(k, z=0, zdir="y")
#==========================================

plt.show()
#-----------------------------------------------------------
label='...'
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()
#p.show()
```

# point class

```python

class Point:
	"""
	Simple class for representing a point in a Cartesian coordinate system.
	"""
	def __init__(self, x, y):
		"""
		Create a new Point at x, y.
		"""
		self.x = x
		self.y = y
	
	def translate(self, dx, dy):
		"""
		Translate the point by dx and dy in the x and y direction.
		"""
		self.x += dx
		self.y += dy
	def __str__(self):
		return("Point at [%.2f, %.2f]" % (self.x, self.y))

p1 = Point(0, 0) # this will invoke the __init__ method in the Point class
print(p1) # this will invoke the __str__ method

p2 = Point(1, 1)
p1.translate(0.25, 1.5)
print(p1)
print(p2)
```

```python
#!/usr/bin/env python

# implement the example graphs/integral from pyx
from pylab import *
from matplotlib.patches import Polygon

def func(x):
    return (x-3)*(x-5)*(x-7)+85

ax = subplot(111)

a, b = 2, 9 # integral area
x = arange(0, 10, 0.01)
y = func(x)
plot(x, y, linewidth=1)

# make the shaded region
ix = arange(a, b, 0.01)
iy = func(ix)
verts = [(a,0)] + list(zip(ix,iy)) + [(b,0)]
poly = Polygon(verts, facecolor='0.8', edgecolor='k')
ax.add_patch(poly)

text(0.5 * (a + b), 30,
     r"$\int_a^b G(x)\mathrm{d}x$", horizontalalignment='center',
     fontsize=20)

axis([0,10, 0, 180])
figtext(0.9, 0.05, 'x')
figtext(0.1, 0.9, 'y')
ax.set_xticks((a,b))
ax.set_xticklabels(('a','b'))
ax.set_yticks([])
show()
```
